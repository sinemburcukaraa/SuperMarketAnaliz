```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['day'] =(df['Date']).dt.day
df['Month'] =(df['Date']).dt.month
df['year'] =(df['Date']).dt.year
```

```python
df['month_name']=df['Date'].dt.month_name()
```

```python
df['weekday']=df['Date'].dt.day_name()
```

```python
df['Time']=pd.to_datetime(df['Time'])
```

```python
df['Hour']=df['Date'].dt.hour
df
```

```python
plt.figure(figsize=(10,6))
plt.title('Monthly transaction')
sns.countplot(df['month_name'])
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(8,6))
plt.title('Male vs Fmale Customers')
sns.countplot(df['Gender']);
```

```python
plt.figure(figsize=(8,8))
explode=(0.1,0.0,0)
df['month_name'].value_counts().plot.pie(autopct='%1.2f%%' ,
                                         startangle=60, explode=explode)
plt.title('Transaction per month')
```

```python
plt.figure(figsize=(8,8))
explode=(0.1,0.0,0,0,0,0)
df['Product line'].value_counts().plot.pie(autopct='%1.2f%%' ,
                                         startangle=60,
                                        explode=explode)
plt.title('Transaction per month')
```
